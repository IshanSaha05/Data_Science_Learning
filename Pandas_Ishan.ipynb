##**Indexing in Pandas**

**Index-Based Selection**

```python
# The iloc is index based indexing.
# It can be used just like slicing operators are used in python.
# The first value selects the row and the second one column.

reviews.iloc[0]
# reviews.iloc[:,0]
# reviews.iloc[:3,0]
# reviews.iloc[1:3,0]
# reviews.iloc[[0,1,2],0]
# reviews.iloc[-5:]
```

**Label-Based Selection**

```python
# The loc is label based indexing.
# It can be used just like slicing operators used in python.
# The first value selects the row and the second one column.
# It can be used just like iloc containing both digits for rows and cols.
# But for the col part it can have col_names or col labels too.
# And for the row part it can have boolean expressions too.

# V.V.Imp: The loc is label based indexing means, if the row index is not sequential like [0,1,2,...], and it is like [5,665,9,7,12,...] random nos, where,
# 5 represents the 1st row, 7 the 2nd and so on, then if we use dataframe.loc[0:3], it wont select any rows, rather will give error. Bcz it never interprets
# the values given in the loc as an integer position along the index. In loc, we need to give the exact row values. Like if we need to select the first 4 rows,
# we need to say dataframe.loc[5:7], or dataframe.loc[:7], where rows with indeces 5,665,9,7 will be selected. Similarly, if we write, dataframe.loc[:665],
# it wouldn't select all the rows from 0 to 665 (even if that no of rows exists in the dataframe), rather it will only select the first 2 rows with indeces 
# 5 & 665. 

reviews.loc[0,'country']
# reviews.loc[:,['taster_name','taster_twitter_handle','points']]
```

**Choosing Between loc and iloc**

```python
# In case of iloc this indexing will choose rows of nos --> 0,1,2,3,....,99

reviews.iloc[0:100]
```

```python
# In case of loc this indexing will choose rows of nos --> 0,1,2,3,....,100

reviews.loc[0:100]
```

# **Grouping & Sorting**

## **Groupwise Analysis**

```python
# The groupby function is used to group large amounts of data and compute operations on these groups.
# We can give axis varibale to it. 0 --> split along rows, 1 --> split along cols.
# The dropna bool variable can be used. If True --> Na are dropped as keys. False --> Na are treated as keys.
# This returns a groupby object and not a series or dataframe. 

reviews.groupby('price').points.count()
# reviews.groupby('points').price.min()
```

```python
# We can use lambda functions also.

reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
# We can use as many as cols to make a groupby object.

reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
# We can also use agg function to use, where we can pass some aggregate functions.
# len gives the number of rows under same groupby row tag.
# min gives the minimum value of a particular chosen col for the groupby row tags.
# Similarly, max gives the maximum.

reviews.groupby(['country']).price.agg([len, min, max])
```

## **Multi-indexes**

```python
# This is a multi-index dataframe, i.e. it is different from regular dataframes bcz it has more than one level to it.

countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
# This shows that the index of the countries_reviewd dataframe is multiindex object.

mi=countries_reviewed.index
type(mi)
```

```python
# Reset index can be used to remove one or more levels from a multiindex dataframe.
# If the index is multiple levels, i.e. multiindex dataframe, we can reset one or more or all of them.
# For that we need to mention those levels inside the level variable. If not mentioned, all of them gets reseted.
# If all the levels are reseted, then the defualt one is used.
# Reset index will change the order of rows of the original dataframe. The arrangement after will be equivalent to the arrangement of the rows due to groupby function.
# It will not return to the original dataframe arrangement of rows before groupby function is used.
# If we donot mention any level all the levels gets reseted automatically.
# The description here is a col name.

reviews.groupby(['country', 'province']).description.agg([len]).reset_index()
```

```python
# The sort_values is used to sort a dataframe along a certain axis.
# In by we give the name of the cols we want to use to sort the values.
# Here ascending parameter can be used.

reviews.groupby(['country', 'province']).description.agg([len]).reset_index().sort_values(by='len',ascending=False)
```

```python
# We use this method to sort the row label or the indeces.
# A parameter kind can be used to tell  what kind of sort algo needs to be used.
# By default it is Quicksort.

reviews.groupby(['country', 'province']).description.agg([len]).reset_index().sort_index()
```

# **Renaming & Combining**

## **Renaming**

```python
# The rename function can be used to rename a particular col name or index name.
# Here, we have renamed a col using dictionary.
# To do that we pass a dictionary to the columns varibale.
# The dictionary key will be the current col name and the value will be the new col name.

reviews.rename(columns={'points': 'score'})
```

```python
# Here, we are renameing the index name.
# Dictionary is passed to the variable index.
# The key represents the previous index name and the value represents the new index name.
# This dictionary got two key-value pairs. So, two index name are being changed at the same time.

reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

```python
# Set the name of the axis for the index or columns.
# The rename_axis is used to either give name or rename a particular axis, i.e. either give name to the index axis or the col axis.
# Here, we have done both.

reviews.rename_axis("wines", axis='rows').rename_axis("fields", axis='columns')
```

## **Combining**

```python
# We can concat csv's using the concat fucntion.
# Here, we have concated two csv's by passing the two of them in a list.
# Concatenate pandas objects along a particular axis with optional set logic along the other axes.

canadian_youtube = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Pandas/CAvideos.csv")
british_youtube = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Pandas/GBvideos.csv")

pd.concat([canadian_youtube, british_youtube])
```

```python
# Join columns of another DataFrame.
# Join columns with other DataFrame either on index or on a key col.
# Efficiently join multiple DataFrame objects by index at once by passing a list.

left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```
